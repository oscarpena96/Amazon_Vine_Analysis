```python
#Import dependencies
import pandas as pd
```

```python
#Import data

data = pd.read_csv("vine_table2.csv")
```

```python
#check data

data.head()
```

```python
# check if datatype is correct 
data.dtypes
```

```python
# check summary
data.describe()
```

```python
#get total votes over 20
total_votes = data[data["total_votes"] >=20]
```

```python
total_votes.head()
```

```python
#filter votes where 50% of total votes where helpful_votes
votes_filtered = total_votes.query("(helpful_votes)/(total_votes ) >=0.5")
votes_filtered.count()
```

```python
#filter vine users
paid_vine = votes_filtered[votes_filtered["vine"]=="Y"]
paid_vine
```

```python
#filter non-vine users
non_vine = votes_filtered[votes_filtered["vine"]=="N"]
non_vine
```

```python
#filter 5 star reviews from vine users
vine_5_star = paid_vine[paid_vine["star_rating"] ==5]
vine_5_star
```

```python
#filter 5 star reviews from non-vine users
non_5_star = non_vine[non_vine["star_rating"] ==5]
non_5_star
```

```python
#get thepercentage of five star vine users
vine_percent = (vine_5_star["star_rating"].count()/paid_vine["star_rating"].count())*100
vine_percent
```

```python
non_vine_percent = (non_5_star["star_rating"].count()/non_vine["star_rating"].count())*100
non_vine_percent
```
